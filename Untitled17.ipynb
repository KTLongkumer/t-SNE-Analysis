```python
file_path = 'Brain_GSE50161.csv'
data = pd.read_csv(file_path)
```

```python
# Handle missing values
data = data.fillna(data.mean())

# Encode categorical variables
categorical_cols = data.select_dtypes(include=['object']).columns
data = pd.get_dummies(data, columns=categorical_cols)

# Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
def plot_tsne(perplexity, learning_rate):
    tsne = TSNE(n_components=2, perplexity=perplexity, learning_rate=learning_rate, random_state=42)
    data_tsne = tsne.fit_transform(data_scaled)
    tsne_df = pd.DataFrame(data_tsne, columns=['TSNE1', 'TSNE2'])
    
    plt.figure(figsize=(10, 8))
    plt.scatter(tsne_df['TSNE1'], tsne_df['TSNE2'])
    plt.title(f't-SNE with Perplexity={perplexity}, Learning Rate={learning_rate}')
    plt.xlabel('TSNE1')
    plt.ylabel('TSNE2')
    plt.show()

# Example: Try different values of perplexity and learning rate
for perplexity in [5, 30, 50]:
    for learning_rate in [10, 100, 200]:
        plot_tsne(perplexity, learning_rate)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Load your data
data = pd.read_csv('Brain_GSE50161.csv')

# Separate features and target
X = data.drop(columns=['samples', 'type'])
y = data['type']

# Preprocess the features (e.g., standardize)
scaler = StandardScaler()
X_preprocessed = scaler.fit_transform(X)

# Encode target labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Initialize t-SNE model
tsne = TSNE(
    n_components=2,
    perplexity=30,
    learning_rate=200,
    n_iter=1000,
    random_state=42
)

# Train the t-SNE model
X_tsne = tsne.fit_transform(X_preprocessed)

# Plot the t-SNE result
plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y_encoded, cmap='viridis')
plt.colorbar(scatter, ticks=range(len(label_encoder.classes_)), format=plt.FuncFormatter(lambda x, pos: label_encoder.classes_[int(x)]))
plt.title('t-SNE Visualization')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.show()
```

```python
from sklearn.manifold import TSNE
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import pairwise_distances
import numpy as np

# Define a grid of hyperparameters
perplexities = [5, 30, 50]
learning_rates = [10, 200, 1000]
n_iter = [500, 1000, 1500]

best_score = float('inf')
best_params = {}

# Grid search
for perplexity in perplexities:
    for learning_rate in learning_rates:
        for n_iter_ in n_iter:
            tsne = TSNE(
                n_components=2,
                perplexity=perplexity,
                learning_rate=learning_rate,
                n_iter=n_iter_,
                random_state=42
            )
            
            # Fit and transform
            X_tsne = tsne.fit_transform(X_preprocessed)
            
            # Evaluate clustering (using a metric like pairwise distances)
            # Calculate mean pairwise distance between points in the reduced space
            mean_dist = np.mean(pairwise_distances(X_tsne))
            
            # Update the best score and parameters
            if mean_dist < best_score:
                best_score = mean_dist
                best_params = {'perplexity': perplexity, 'learning_rate': learning_rate, 'n_iter': n_iter_}

print("Best Parameters:", best_params)
print("Best Score (Mean Pairwise Distance):", best_score)
```

```python
from sklearn.manifold import TSNE
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import pairwise_distances
import numpy as np
import random

# Define ranges for hyperparameters
perplexities = [5, 30, 50]
learning_rates = [10, 200, 1000]
n_iter = [500, 1000, 1500]

# Random search
best_score = float('inf')
best_params = {}

for _ in range(10):  # Number of random samples
    perplexity = random.choice(perplexities)
    learning_rate = random.choice(learning_rates)
    n_iter_ = random.choice(n_iter)
    
    tsne = TSNE(
        n_components=2,
        perplexity=perplexity,
        learning_rate=learning_rate,
        n_iter=n_iter_,
        random_state=42
    )
    
    # Fit and transform
    X_tsne = tsne.fit_transform(X_preprocessed)
    
    # Evaluate clustering
    mean_dist = np.mean(pairwise_distances(X_tsne))
    
    # Update the best score and parameters
    if mean_dist < best_score:
        best_score = mean_dist
        best_params = {'perplexity': perplexity, 'learning_rate': learning_rate, 'n_iter': n_iter_}

print("Best Parameters:", best_params)
print("Best Score (Mean Pairwise Distance):", best_score)
```
